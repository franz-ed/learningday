## importing python modules

For heavy numeric compuation there is a python liberary called "numpy". We will import this library and give it the alias name "np":

```python
import numpy as np
```

For heavy calculation we use vectors (a list of numbers) and matrices (an array of numbers). numpy can deal very well with such arrays. e.g. we use the np.random.rand function to create a 3 by 3 matrix of random numbers between 0 and 1:

```python
np.random.rand(3,3)
```

of course we can also store these vectors and marices in a varaible. E.g. let us create a vector consisting of the numbers 0 to 9 and 10 by 10 matrix with all 1s:

```python
v=np.arange(10)
A=np.ones((10,10))
A
```

```python
v
```

Variable names in python can be any string and it is also case sensitive. Here we will use Upper Case for a matrix and lower case for a vector. (But this is just a convention here). We can multiply a matrix with a vector

```python
A*v
```

This might be not what you expected: Python tries something that is called "broadcast". It sees that our vector is length then and the matrix also has 10 as nn inner dimension. So it it multiplies every row with this vector. If we want the "real" multiplication of a matrix times a vector we need to use the "dot product":

```python
A.dot(v)
```

```python
I=np.eye(10)
I
```

The "eye" matrix consisting only of diagonal elements of one will transform a vector onto itself. Instead of writing I.dot(v) we can also use the @ sign for the dot product: I @ v

```python
I @ v
```

The dot product of a vector with itself is the length of the vector squared:

```python
v.dot(v)
```

```python
v @ v * 1.0
```
